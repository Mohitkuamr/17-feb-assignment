```python
Ans = 1
MongoDB is a popular document-oriented NoSQL database system that is designed to handle unstructured data.
It stores data in flexible, JSON-like documents, allowing for a more dynamic and expressive data model 
compared to traditional relational databases.

Non-relational databases, also known as NoSQL databases, are a type of database management system
that does not use the traditional tabular relational database model of rows and columns. Instead, 
non-relational databases use a variety of data models, such as key-value, document-oriented, or graph-based,
to store and manage data.
```

```python
Ans = 2
1.
Document-oriented: MongoDB is a document-oriented database, which means that it stores data in flexible,
JSON-like documents rather than traditional tables with rows and columns. This makes it easy to handle 
unstructured and semi-structured data, and allows for more expressive data models.

2.
High availability and scalability: MongoDB is designed to be highly available and scalable,
making it a good choice for applications with high traffic and large volumes of data. It supports 
automatic sharding and replica sets for automatic failover and load balancing.

3. 
Rich query language: MongoDB provides a rich query language with support for ad-hoc queries, text search, 
and aggregation operations. This allows for complex queries to be performed easily and efficiently.

4.
Indexing: MongoDB supports a variety of indexing options, including compound indexes, text indexes,
and geospatial indexes. This allows for fast query performance even on large data sets.

5.
Schema flexibility: MongoDBs dynamic schema allows for easy schema evolution without requiring downtime
or complex migration scripts. This makes it ideal for agile development environments where requirements 
may change rapidly.

6.
Built-in aggregation framework: MongoDBs built-in aggregation framework allows for complex data 
transformations and analytics to be performed within the database itself. This can significantly reduce
the amount of data that needs to be transferred to other systems for analysis.

7.
Automatic sharding and load balancing: MongoDB automatically shards data across multiple servers,
allowing for horizontal scaling and improved performance. It also supports automatic load balancing, 
making it easy to add or remove servers as needed.
```

```python
Ans = 4
import pymongo

# Connect to the MongoDB client running on localhost
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create a new database called "mydatabase"
mydb = client["mydatabase"]

# Create a new collection called "customers" in the "mydatabase" database
mycol = mydb["customers"]

# Insert one record into the "customers" collection
record = {"name": "John", "address": "Highway 37"}
mycol.insert_one(record)

# Insert multiple records into the "customers" collection
records = [
  {"name": "Peter", "address": "Lowstreet 27"},
  {"name": "Amy", "address": "Apple st 652"},
  {"name": "Hannah", "address": "Mountain 21"},
  {"name": "Michael", "address": "Valley 345"},
  {"name": "Sandy", "address": "Ocean blvd 2"},
  {"name": "Betty", "address": "Green Grass 1"},
  {"name": "Richard", "address": "Sky st 331"}
]
mycol.insert_many(records)

# Print the inserted record using find_one() method
print(mycol.find_one({"name": "John"}))

# Print all the inserted records using find() method
for record in mycol.find():
    print(record)
```
